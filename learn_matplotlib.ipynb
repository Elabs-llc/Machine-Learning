## Matplotlib Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
# prepare data
x = [1,2,3,4]
y = [11, 22, 33,44]

# setup plot
fig,ax = plt.subplots(figsize=(5,5))

# plot data
ax.plot(x,y)

# customize plot
ax.set(title="A Simple Plot Data", xbound="x-axis", ylabel="y-axis")

# save and show
fig.savefig('simple-plot.png')
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
## oo method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="thalach",
             ax=ax,
             colorbar="target")
ax.set(title="People Ages - Thalach");
```

```python
fix, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(
             x=over_50["age"],
             y=over_50["chol"],
             c=over_50["target"])
ax.set(title="People Ages - Cholestrol", xlabel="Age", ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(y=over_50["chol"].mean(), linestyle="--");
```
